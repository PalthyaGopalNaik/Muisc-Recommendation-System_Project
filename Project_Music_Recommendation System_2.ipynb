```python
# Checking The Outliers

song_2= song_1[song_1.genre.isin(song_1.genre.value_counts().index)]
sns.boxplot(x='genre',y='release_year',data=song_2)
plt.show()

```

```python
# Removing The Outliers

# Define a function to remove outliers based on IQR
def remove_outliers(df, column, threshold=1.5):
    q1 = df[column].quantile(0.25)
    q2 = df[column].quantile(0.75)
    iqr = q2 - q1
    lower_bound = q1 - threshold * iqr
    upper_bound = q2 + threshold * iqr
    df_no_outliers = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df_no_outliers

# Remove outliers for each genre
song_2_no_outliers = song_2.groupby('genre').apply(lambda x: remove_outliers(x, 'release_year'))

# Flatten the grouped DataFrame
song_2_no_outliers = song_2_no_outliers.reset_index(drop=True)

```

```python
# Create a boxplot without outliers

plt.figure(figsize=(12, 8))
sns.boxplot(x='genre', y='release_year', data=song_2_no_outliers, showfliers=False)
plt.title('Boxplot of Release Year for Each Genre (No Outliers)')
plt.show()
```

# Model Building

```python
pivot_table = pd.pivot_table(song_2, values=['danceability', 'loudness', 'acousticness', 'instrumentalness', 'valence', 'energy'], index=['songs','artists','genre','lyrics'], columns='release_year', fill_value=0)
pivot_table.head()
```

```python
# Sparse matrices are especially useful when dealing with large datasets in machine learning, where most of the features might be zero for each sample.
from scipy.sparse import csr_matrix

song_features_df_matrix = csr_matrix(pivot_table.values)
song_features_df_matrix
```

```python
from sklearn.neighbors import NearestNeighbors


model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'auto')
model_knn.fit(song_features_df_matrix)
```

```python
# Parameter Tunning
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import GridSearchCV
param_grid = {'n_neighbors': [5, 10, 15, 20]}
model_knn = NearestNeighbors(metric='cosine', algorithm='auto')
grid_search = GridSearchCV(model_knn, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(song_features_df_matrix)
print("Best Parameters:", grid_search.best_params_)

```

```python
# Example: Get recommendations for a specific song name
song_name = '"african head charge"'  # Replace 'Your_Song_Name' with the name of the song you want recommendations for

# Locate the index of the song in the pivot_table DataFrame
song_index = pivot_table.index.get_loc(song_name)

# Retrieve the data corresponding to the specified index
selected_data = pivot_table.iloc[song_index, :]

# Print the selected data
print("Selected Data:")
print(selected_data)

# Query the nearest neighbors for the specified index
distances, indices = model_knn.kneighbors(selected_data.values.reshape(1, -1), n_neighbors=5)
```

```python
print('Recommendations for {0}:\n'.format(song_name))
for i in range(0, len(distances.flatten())):
    if i == 0:
        continue  # Skip the first neighbor, as it will be the query point itself
    print('{0}: {1}, with distance of {2}'.format(i, pivot_table.index[indices.flatten()[i]], distances.flatten()[i]))
```

```python
# Create a DataFrame to display the results
Neighbors_Df = pd.DataFrame({'Neighbor Index': indices.flatten(),'Distance': distances.flatten()})
# Fetch additional information from pivot_table using neighbor indices
Neighbors_Df['Songs'] = pivot_table.index[Neighbors_Df['Neighbor Index']].get_level_values('songs').tolist()
Neighbors_Df['Artists'] = pivot_table.index[Neighbors_Df['Neighbor Index']].get_level_values('artists').tolist()
Neighbors_Df['Genre'] = pivot_table.index[Neighbors_Df['Neighbor Index']].get_level_values('genre').tolist()
Neighbors_Df['lyrics'] = pivot_table.index[Neighbors_Df['Neighbor Index']].get_level_values('lyrics').tolist()
# Rearrange the columns
Neighbors_Df = Neighbors_Df[['Songs', 'Artists','Genre','lyrics','Neighbor Index', 'Distance']]


# Sort the DataFrame by the 'Distance' column in ascending order
Neighbors_Df = Neighbors_Df.sort_values(by='Distance')
Neighbors_Df.head()
```
